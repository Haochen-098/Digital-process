```
#Your Code 
from skimage import io
from skimage import filters

img = io.imread('noisy.jpg')
gau_img = filters.gaussian(img)
print("Orignal image")
io.imshow(img) 
io.show()
print("Gaussian filter")
io.imshow(gau_img) 
io.show()
print("median filter")
med_img = filters.median(img)
io.imshow(med_img) 
io.show()
print( "median filter was more successful.")
```
