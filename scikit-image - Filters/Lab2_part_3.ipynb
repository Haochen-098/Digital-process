```
#Your Code 
from skimage import io
from skimage import filters

from skimage import io
img = io.imread('damage_cameraman.png')
mask = io.imread('damage_mask.png')
gau_img = filters.gaussian(img)
for i in range (0,15):
  for x in range(0,img.shape[0]):
    for y in range(0,img.shape[1]):
      if mask[x][y] != 0:
        gau_img[x][y] = img[x][y]
  gau_img = filters.gaussian(gau_img)
io.imshow(gau_img) 
io.show()
```
