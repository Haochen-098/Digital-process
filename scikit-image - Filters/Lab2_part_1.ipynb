```
#Your Code
from skimage import io
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img = io.imread('moon.png',True)  
img_1= np.zeros(img.shape)
img_2 = np.zeros(img.shape)
img_3 = np.zeros(img.shape)
img_4 = np.zeros(img.shape)
lap = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
mat_2 = np.array([[0,0,0],[0,1,0],[0,0,0]])
mat_3 = np.array([[0,0,0],[0,0,1],[0,0,0]])
mat_4 = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
height = img.shape[0]
width = img.shape[1]
for x in range(0,height-1):
  for y in range(0,width-1):
    sum_1 = 0
    sum_2 = 0
    sum_3 = 0
    sum_4 = 0
    for a in range(-1,2):
      for b in range(-1,2):
        sum_1 += int(lap[a+1,b+1]) * int(img[x+a,y+b])
        sum_2 += int(mat_2[a+1,b+1]) * int(img[x+a,y+b])
        sum_3 += int(mat_3[a+1,b+1]) * int(img[x+a,y+b])
        sum_4 += int(mat_4[a+1,b+1]) * int(img[x+a,y+b])     
    img_1[x,y]= sum_1
    img_2[x,y] = sum_2
    img_3[x,y]= sum_3
    img_4[x,y]= sum_4
print("Laplacian Filter")
img_1 = rgb2gray(img_1)
plt.imshow(img_1,cmap='gray') 
plt.show()
print("Matrix 2 Filter")
img_2 = rgb2gray(img_2)
plt.imshow(img_2,cmap='gray')
plt.show()
print("Matrix 3 Filter")
plt.imshow(img_3,cmap='gray')  
plt.show()
print("Mean Filter")
plt.imshow(img+(img-img_4),cmap='gray') 
plt.show()
```
