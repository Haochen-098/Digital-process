```
from skimage import io
from skimage.filters import sobel_h 
from skimage.filters import sobel_v
from skimage.filters import gaussian
from scipy import ndimage as ndi
from skimage import feature
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img = io.imread('ex2.jpg',True) 
plt.imshow(img,cmap='gray') 
plt.title("Original")
plt.show()
img_g = gaussian(img)
plt.imshow(img_g,cmap='gray') 
plt.title("Gaussian filtering")
plt.show()

#1
img_1 = feature.canny(img,sigma=1.0, low_threshold = 25)
plt.subplot(2,2,1)
plt.imshow(img_1,cmap='gray') 
plt.title("low threshold = 25")
img_2 = feature.canny(img,sigma=1.0, low_threshold = 50)
plt.subplot(2,2,2)
plt.imshow(img_2,cmap='gray') 
plt.title("low threshold = 50")
img_3 = feature.canny(img,sigma=1.0, high_threshold = 150)
plt.subplot(2,2,3)
plt.imshow(img_3,cmap='gray') 
plt.title("high threshold = 150")
img_4 = feature.canny(img,sigma=1.0, high_threshold = 200)
plt.subplot(2,2,4)
plt.imshow(img_4,cmap='gray') 
plt.title("high threshold = 200")
plt.suptitle("fix sigma = 1.0")
plt.show()

#2
img_5 = feature.canny(img,sigma=1.0, low_threshold=50,high_threshold=150)
plt.subplot(2,2,1)
plt.imshow(img_5,cmap='gray') 
plt.title("sigma = 1.0")
img_6 = feature.canny(img,sigma=1.5, low_threshold=50,high_threshold=150)
plt.subplot(2,2,2)
plt.imshow(img_6,cmap='gray') 
plt.title("sigma = 1.5")
img_7 = feature.canny(img,sigma=2.0, low_threshold=50,high_threshold=150)
plt.subplot(2,2,3)
plt.imshow(img_7,cmap='gray') 
plt.title("sigma = 2.0")
img_8 = feature.canny(img,sigma=2.5, low_threshold=50,high_threshold=150)
plt.subplot(2,2,4)
plt.imshow(img_8,cmap='gray') 
plt.title("sigma = 2.5")
plt.suptitle(" fix low_threshold=50 and high_threshold=150")
plt.show()
```

Apply the conny will convert the pixel value lower than the set value to black(0),the higher value to white(225). The threshold value become larger, the less white part reserve.
Sigma is the Standard deviation of the Gaussian filter. Larger sigma, the lower Gaussian filter impulse response, more part convert to black(0) by applying threshold value.
