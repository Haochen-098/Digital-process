```
import numpy as np
import skimage.exposure
import skimage.io
import matplotlib.pyplot as plt


def own_code(image):
    own_hist= np.zeros(256)
    for axes_x in image:
        for axes_y in axes_x:
           own_hist[axes_y] += 1
    return own_hist

def main():
    image = skimage.io.imread('test.jpg')
    histor = plt.subplot(1,2,1)
    histor.set_title('Original image')
    plt.imshow(image)
    plt.xlim([0,256])
    plt.show()
    own_hist = own_code(image)
    histor=plt.subplot(1,2,2)
    histor.set_title('Histogram')
    plt.plot(own_hist)
    plt.xlim([0,256])
    plt.show()
    equ_image = image.copy()
    dim = list(image.shape)
    x = dim[0]
    y = dim[1]
    for a in range(0, x):
        for b in range(0, y):
          equ_image[a,b] = int((255/int(x*y))*(own_hist[a])+0.5)
    histor = plt.subplot(2,2,1)
    histor.set_title('New image')
    plt.imshow(equ_image)
    plt.xlim([0,256])
    plt.show()
    equ_hist = own_code(equ_image)
    histor=plt.subplot(2,2,2)
    histor.set_title('Equalization histogram')
    plt.plot(equ_hist)
    plt.xlim([0,256])
    plt.show()

   

main()
```
