```
#Your Code 
import numpy as np
import skimage.exposure
import skimage.io
import matplotlib.pyplot as plt

def own_code(image):
    own_hist= np.zeros(256)
    for axes_x in image:
        for axes_y in axes_x:
           own_hist[axes_y] += 1
    return own_hist

def main():
    image = skimage.io.imread('test.jpg')
    own_hist = own_code(image)
    ski_hist,bins = skimage.exposure.histogram(image,nbins=256)
    histor=plt.subplot(1,2,1)
    histor.set_title('Histogram by Own code')
    plt.plot(own_hist)
    plt.xlim([0,256])
    histor=plt.subplot(1,2,2)
    histor.set_title('Histogram by Skimage')
    plt.plot(bins,ski_hist)
    plt.xlim([0,256])
    plt.show()
main()
```
