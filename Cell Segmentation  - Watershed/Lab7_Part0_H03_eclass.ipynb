**Explain the role of the compactness parameter. Justify the outputs that you got in previous part.**

Under the n_segments = 5, the (approximate) number of labels in the segmented output image are 5. Compactness parameter is the balances color proximity and space proximity. Higher values give more weight to space proximity, making superpixel shapes more square/cubic. So as the compactness value incress, the contrast become stronger and more obvious color block distinction.
